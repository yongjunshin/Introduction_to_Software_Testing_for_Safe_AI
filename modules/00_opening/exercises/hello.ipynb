```python
# Simple Library Import Test with Versions
# This cell tests if all required packages can be imported and shows their versions

print("Testing library imports and versions...")
print("=" * 50)

# Each entry is (display name, import name) from env/requirements.txt
packages = [
    ('torch', 'torch'),
    ('torchvision', 'torchvision'),
    ('torchinfo', 'torchinfo'),
    ('numpy', 'numpy'),
    ('pandas', 'pandas'),
    ('matplotlib', 'matplotlib'),
    ('sklearn', 'sklearn'),
    ('jupyterlab', 'jupyterlab'),
    ('ipywidgets', 'ipywidgets'),
    ('httpx', 'httpx'),
    ('pytest', 'pytest'),
    ('pytest-cov', 'pytest_cov'),
    ('coverage', 'coverage'),
    ('tabulate', 'tabulate'),
    ('mutmut', 'mutmut'),
    ('z3-solver', 'z3'),
    ('shapely', 'shapely'),
]


def get_version(module):
    # Most packages expose __version__; z3 only has get_version_string()
    version = getattr(module, '__version__', None)
    if version is None and hasattr(module, 'get_version_string'):
        version = module.get_version_string()
    return version or 'unknown'


for display_name, import_name in packages:
    try:
        module = __import__(import_name)
        print(f"✓ {display_name:15} | version: {get_version(module)}")
    except ImportError as e:
        print(f"✗ {display_name:15} | ERROR: {e}")

print("=" * 50)
print("Import test complete!")
```
