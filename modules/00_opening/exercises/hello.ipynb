```python
# Simple Library Import Test with Versions
# This cell tests if all required packages can be imported and shows their versions

print("Testing library imports and versions...")
print("=" * 50)

# Test each package from requirements.txt
packages = ['torch', 'torchvision', 'numpy', 'pandas', 'matplotlib', 'sklearn', 'jupyterlab', 'ipywidgets', 'httpx', 'pytest', 'coverage']

for package in packages:
    try:
        module = __import__(package)
        version = getattr(module, '__version__', 'unknown')
        print(f"✓ {package:15} | version: {version}")
    except ImportError as e:
        print(f"✗ {package:15} | ERROR: {e}")

print("=" * 50)
print("Import test complete!")
```
