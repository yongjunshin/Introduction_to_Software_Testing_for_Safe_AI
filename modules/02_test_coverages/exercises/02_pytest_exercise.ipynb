## Running PyTest

`PyTest` automatically discovers test files and functions based on naming conventions.
It runs tests by looking for: 
- Files named `test_*.py` or `*_test.py`
- Functions prefixed with `test_` inside those files

It searches the current directory and all subdirectories.

### In This Exercise

Running `pytest` will execute all three `test_*` functions defined in `test_classify_triangle.py`.

Run the three tests by executing the following command:

````python
!pytest
````

## Measuring Statement and Branch Coverage with PyTest

The plain `pytest` command only runs tests but does not collect coverage data. To measure coverage, we use the `pytest-cov` plugin.

### Statement Coverage

Use the `--cov` option to measure **statement coverage**:

```bash
pytest --cov=<module_name>
```

| Column | Description |
|--------|-------------|
| **Stmts** | Total number of executable statements |
| **Miss** | Number of statements not executed |
| **Cover** | Percentage of statements executed |

Run the following command to measure statement coverage for `classify_triangle.py`:

````python
!pytest --cov=classify_triangle
````

### Branch Coverage

To also measure **branch coverage** (ensuring every `if`/`else` path is tested), add the `--cov-branch` option:

```bash
pytest --cov=<module_name> --cov-branch
```

This adds two more columns to the output:

| Column | Description |
|--------|-------------|
| **Branch** | Total number of branches in the code |
| **BrPart** | Number of branches only partially covered |
| **Cover** | Combined coverage (%) calculated as:  `((Stmts-Miss) + (Branch-BrPart)) / (Stmts + Branch) * 100` |

Run the following command to measure both statement and branch coverage:

````python
!pytest --cov=classify_triangle --cov-branch
````

## Generating an HTML Coverage Report

The terminal output provides a quick summary, but for detailed analysis, you can generate an **interactive HTML report**.

```bash
coverage html
```

Run the following command.

````python
!coverage html
````

This creates a `htmlcov/` directory containing:
- `index.html` — Overview of all files with coverage percentages
- Individual file pages — Line-by-line highlighting showing:
  - ✅ **Green**: Executed lines
  - ❌ **Red**: Missed lines
  - 🟡 **Yellow**: Partially covered branches

Open `htmlcov/index.html` in a browser to explore the detailed report.

## 🛠️ Running the `coverage` Tool Directly

So far, we've used `pytest-cov` to measure coverage through test functions. But what if you want to measure coverage by **running a script directly** with specific inputs?

The `coverage` command-line tool allows you to:
- Run any Python script while tracking coverage
- Accumulate coverage data across multiple runs
- Generate reports without writing test functions

### Key Commands

| Command | Description |
|---------|-------------|
| `coverage erase` | Clear previous coverage data |
| `coverage run script.py [args]` | Run a script and collect coverage |
| `coverage run -a script.py [args]` | Append coverage data to existing data |
| `coverage report` | Display coverage summary in terminal |
| `coverage html` | Generate detailed HTML report |

### Statement Coverage

Try running `classify_triangle.py` directly with different inputs and observe how statement coverage accumulates:

````python
# 1) 이전 커버리지 데이터 삭제
!coverage erase

# 2) 여러 입력으로 스크립트 실행 (append 모드 -a)
!coverage run classify_triangle.py 3 3 3
!coverage run -a classify_triangle.py 3 4 5
!coverage run -a classify_triangle.py 4 5 6

# 3) 리포트 보기
!coverage report
!coverage html
````

### Branch Coverage

To track branch coverage, add the `--branch` flag. Try running with different inputs and observe how branch coverage accumulates:

````python
# 1) 이전 커버리지 데이터 삭제
!coverage erase

# 2) 여러 입력으로 스크립트 실행 (append 모드 -a, --branch 플래그 추가)
!coverage run --branch classify_triangle.py 3 3 3
!coverage run -a --branch classify_triangle.py 3 4 5
!coverage run -a --branch classify_triangle.py 4 5 6

# 3) 리포트 보기
!coverage report
!coverage html
````

Check out the HTML report in the `htmlcov/` directory.
