# Neuron Coverage Exercise

In this exercise, we'll learn how to test a DNN by measuring neuron coverages.

We'll practice measuring:
- Neuron Coverage
- K-Multisection Neuron Coverage
- Neuron Boundary Coverage
- Top-k Neuron Coverage
- Top-k Neuron Pattern Coverage

## Software under Test

In this exercise, we'll test a simple **DNN** trained on the **MNIST dataset**.

### About MNIST

The MNIST dataset consists of:
- **60,000 training images** and **10,000 test images**
- Handwritten digits (0–9)
- Grayscale images of size **28×28 pixels**

### Model Architecture

We use a minimal MLP to keep the exercise simple and fast:

| Layer | Neurons | Activation |
|-------|---------|------------|
| Input | 784 | — |
| Dense | 32 | ReLU |
| Output | 10 | Softmax |

**Total neurons: 42** (32 hidden + 10 output)

This tiny model is sufficient to demonstrate all neuron coverage metrics while keeping computation minimal.
