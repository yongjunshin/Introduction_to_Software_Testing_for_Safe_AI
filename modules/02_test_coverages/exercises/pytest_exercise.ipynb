# Pytest Exercise
 
In this exercise, we'll get hands-on experience using `PyTest`, a software testing tool for Python.

We'll practice:
 - Testing an SuT and obtaining reports on **statement coverage** and **branch coverage**.

```python
!pytest
```

```python
!pytest --cov=classify_triangle
```

```python
!pytest --cov=classify_triangle --cov-branch
```

```python
!coverage html
```

```python
# 1) 이전 커버리지 데이터 삭제
!coverage erase

# 2) 여러 입력으로 스크립트 실행 (append 모드 -a)
!coverage run classify_triangle.py 3 3 3
!coverage run -a classify_triangle.py 3 4 5
!coverage run -a classify_triangle.py 4 5 6

# 3) 리포트 보기
!coverage report
!coverage html
```
